# Approach 1 (Best Solution).
- Time Complexity - O(n).
- Space Complexity - O(1).

- Loop i from 0 - 8.
- Each time validate ith row, ith column and ith grid.
- Do it for other rows, column and grid.

```java
class Solution {

  public boolean isValidSudoku(char[][] board) {
    for (int i = 0; i < 9; i++) {
      if (
        !checkRow(i, board) || !checkColumn(i, board) || !checkGrid(i, board)
      ) {
        return false;
      }
    }
    return true;
  }

  public boolean checkRow(int rowId, char[][] board) {
    boolean[] arr = new boolean[9];

    for (int i = 0; i < 9; i++) {
      char val = board[rowId][i];
      if (val == '.') {
        continue;
      }

      if (arr[val - '1'] == true) {
        return false;
      }

      arr[val - '1'] = true;
    }
    return true;
  }

  public boolean checkColumn(int columnId, char[][] board) {
    boolean[] arr = new boolean[9];

    for (int i = 0; i < 9; i++) {
      char val = board[i][columnId];

      if (val == '.') {
        continue;
      }

      if (arr[val - '1'] == true) {
        return false;
      }

      arr[val - '1'] = true;
    }
    return true;
  }

  public boolean checkGrid(int gridId, char[][] board) {
    boolean[] arr = new boolean[9];

    int rowStart = (gridId % 3) * 3;
    int rowEnd = rowStart + 3;
    int colStart = (gridId / 3) * 3;
    int colEnd = colStart + 3;

    for (int i = rowStart; i < rowEnd; i++) {
      for (int j = colStart; j < colEnd; j++) {
        char val = board[i][j];

        if (val == '.') {
          continue;
        }

        if (arr[val - '1'] == true) {
          return false;
        }

        arr[val - '1'] = true;
      }
    }
    return true;
  }
}
```
