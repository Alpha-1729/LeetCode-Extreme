# Approach 1 (Best Solution).
- Time Complexity - O(n).
- Space Complexity - O(1).

1. **Initialize Result Variable**:
   - Initialize an integer `res` to `0`. This will store the cumulative result of the XOR operation.

2. **Determine the Limit**:
   - Calculate the limit value of the sequence as `limit = start + 2 * (n - 1)`. This represents the last element in the sequence:
     - The sequence of numbers is generated by adding `2` to the previous value starting from `start`. Hence, the last number is `start + 2 * (n - 1)`.

3. **Iterate Over the Sequence**:
   - Use a `for` loop to iterate through the sequence of numbers starting from `start` up to `limit`, incrementing by `2` in each iteration.
   - For each number `i` in the sequence, perform the XOR operation with `res` (`res ^= i`).

4. **Return Result**:
   - After completing the XOR operation for all numbers in the sequence, return the final result stored in `res`.


```java
class Solution {
    public int xorOperation(int n, int start) {
        int res = 0;
        int limit = start + 2 * (n - 1);
        for (int i = start; i <= limit; i += 2) {
            res ^= i;
        }
        return res;
    }
}
```

```java
Solution solution = new Solution();
int n = 5;
int start = 0;
int result = solution.xorOperation(n, start);
System.out.println(result);
```
