# Approach 1 (Best Solution using two for loop).
- Time Complexity - O(n).
- Space Complexity - O(1)

- **Variables**:
  - `sum` is used to store the minimum sum found in the first loop and the final maximum absolute sum at the end.
  - `currentSum` tracks the running sum of the subarray as we iterate through `nums`.

- **Finding the Minimum Value**:
  - The first loop iterates through each element in `nums` to find the minimum subarray sum.
  - For each element `num`:
    - `currentSum` is updated by adding `num`.
    - If `num` is less than `currentSum`, `currentSum` is reset to `num`. This step ensures that we start a new subarray at `num` if it results in a smaller sum.
    - `sum` is updated to store the smallest subarray sum found so far.

- **Finding the Maximum Absolute Value**:
  - `sum` is then set to the absolute value of the smallest sum found, and `currentSum` is reset to 0.
  - The second loop finds the maximum subarray sum in a similar manner.
  - For each element `num`:
    - `currentSum` is updated by adding `num`.
    - If `num` is greater than `currentSum`, `currentSum` is reset to `num`.
    - If `currentSum` is greater than `sum`, `sum` is updated to `currentSum`.

- **Final Return**:
  - After iterating through the array twice, the method returns `sum`, which now contains the maximum absolute sum of any subarray in `nums`.



```java
class Solution {
    public int maxAbsoluteSum(int[] nums) {
        int sum = 0;
        int currentSum = 0;

        // Find the min value.
        for (int num : nums) {
            currentSum += num;
            if (num < currentSum)
            {
                currentSum = num;
            }

            if (currentSum < sum) {
                sum = currentSum;
            }
        }

        sum = Math.abs(sum);
        currentSum = 0;
        // Find the max value.
        for (int num : nums) {
            currentSum += num;
            if (num > currentSum)
            {
                currentSum = num;
            }

            if (currentSum > sum) {
                sum = currentSum;
            }
        }

        return sum;;
    }
}
```

# Approach 2 (Best Solution using single for loop).
- Time Complexity - O(n).
- Space Complexity - O(1).

- **Initial Setup**:
  - Four variables are initialized:
    - `minSum` and `maxSum` store the minimum and maximum subarray sums found so far, initialized to the first element of `nums`.
    - `currentMinSum` and `currentMaxSum` track the running minimum and maximum subarray sums ending at the current element.

- **Iterating Through the Array**:
  - The loop starts from the second element (index 1) of `nums`.
  - For each element `num` in the array:
    - `currentMaxSum` and `currentMinSum` are updated by adding the current element to the running sums.
    - If `currentMaxSum` is less than the current element `num`, it is reset to `num` (this step handles the case where starting a new subarray at the current index gives a higher sum).
    - Similarly, if `currentMinSum` is greater than `num`, it is reset to `num`.
    - The variables `minSum` and `maxSum` are updated to keep track of the smallest and largest subarray sums encountered so far.

- **Final Calculation**:
  - After iterating through the array, the method returns the maximum of the absolute values of `minSum` and `maxSum`. This gives the maximum absolute subarray sum.



```java
class Solution {
    public int maxAbsoluteSum(int[] nums) {
        int minSum = nums[0];
        int maxSum = nums[0];

        int currentMinSum = nums[0];
        int currentMaxSum = nums[0];
        for (int i = 1; i < nums.length; i++) {
            int num = nums[i];
            currentMaxSum += num;
            currentMinSum += num;

            if (currentMaxSum < num) {
                currentMaxSum = num;
            }

            if (currentMinSum > num) {
                currentMinSum = num;
            }

            if (currentMinSum < minSum) {
                minSum = currentMinSum;
            }

            if(currentMaxSum > maxSum) {
                maxSum = currentMaxSum;
            }
        }
        return Math.max(Math.abs(minSum), Math.abs(maxSum));
    }
}
```

```java
Solution solution = new Solution();
int[] nums = {1, -3, 2, 3, -4};
int result = solution.maxAbsoluteSum(nums);
System.out.println(result);
```
